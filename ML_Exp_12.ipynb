```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Time series data
time = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
values = np.array([10, 12, 15, 18, 20])

# Create model
model = LinearRegression()

# Train model
model.fit(time, values)

# Predict next value (t=6)
next_time = np.array([[6]])
prediction = model.predict(next_time)
print("Predicted Next Value:", prediction[0])

# Plot graph

plt.plot(time, values, marker='o', label='Original Data')
plt.scatter(6, prediction, label='Predicted Point')
plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Time Series Prediction")
plt.legend()
plt.show()
```
