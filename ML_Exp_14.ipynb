```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# Dataset
data = {
'Movie1': [5, 4, 1, 1, 0],
'Movie2': [3, 0, 1, 0, 1],
'Movie3': [0, 0, 0, 0, 5],
'Movie4': [1, 1, 5, 4, 4]
}
df = pd.DataFrame(data, index=['User1','User2','User3','User4','User5'])

# Calculate similarity
similarity = cosine_similarity(df)
similarity_df = pd.DataFrame(similarity,
index=df.index,
columns=df.index)
print("User Similarity Matrix:\n")
print(similarity_df)

# Recommend for User1
user = 'User1'
similar_users = similarity_df[user].sort_values(ascending=False)[1:2]
print("\nMost Similar User to", user, ":\n", similar_users)
```
