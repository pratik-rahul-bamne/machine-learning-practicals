```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Independent variables: Hours, Attendance, Previous Marks
X = np.array([
[2, 75, 60],
[4, 80, 65],
[6, 85, 70],
[8, 90, 80],
[10, 95, 85]
])
# Dependent variable: Final Marks
y = np.array([55, 60, 70, 85, 95])

# Create model
model = LinearRegression()

# Train model
model.fit(X, y)

# Prediction
prediction = model.predict([[7, 88, 75]])

print("Predicted Final Marks:", prediction[0])
# Model parameters

print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)

# Model accuracy
print("R^2 Score:", model.score(X, y))
```
